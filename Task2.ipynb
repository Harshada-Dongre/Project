# Prediction using Supervised ML

In this task we are using Python Scikit-Learn library for machine learning,to implement regression functions.In this project we are dealing with simple linear regression involving two variables.

## Task

What will be predicted score if a student study for 9.25 hrs in a day?

### Simple Linear Regression

In this regression task we will predict percentage of marks obtained by student based on number of hours they have studied.

### Importing required Libraries

```python
# importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

#### Importing Dataset

```python
file="http://bit.ly/w-data"
data=pd.read_csv(file)
```

#### Representing Data graphically

```python
data.plot(x='Hours', y='Scores', style='o')
plt.title('Time(Hours) vs Score(Percentage)')
plt.xlabel('Time in hours')
plt.ylabel('Score in Percentage ')
plt.show()
```

From the above graph, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.



#### Dividing Data 

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
```


Now that we have our attributes and labels, the next step is to split this data into training and test sets. We'll do this by using Scikit-Learn's built-in train_test_split() method:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

We have the training and testing sets ready for training our model.

### Training the Algorithm

We have split our data into training and testing sets.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

#### Step to make Predictions

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
compare= pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
compare
```

```python
hours = 9.25
own_pred = regressor.predict([[hours]])
print("Count of hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

###### The predicted score if a student studies for 9.25 hrs/day is about 93.69.

### Evaluating the model

```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-2:', metrics.r2_score(y_test, y_pred))
```

###### Thus,from the above result we can see the model's accuracy is of 94.54%
